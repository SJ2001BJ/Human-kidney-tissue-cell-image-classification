## Differentiate between training set and validation set

```python
import keras
import numpy as np

from numpy.random import seed
import tensorflow as tf

import keras.backend as K
import tensorflow as tf
from matplotlib import pyplot as plt

from keras.models import Sequential
from keras.layers import Dense, Conv2D, Flatten,Dropout,MaxPooling2D

from numpy.ma.core import maximum
# Your code here: load the datasets, create the model, train it and then set predictions as usual
from tensorflow.keras.utils import image_dataset_from_directory

color_mode= 'grayscale'
image_size= (28,28)
batch_size = 64
train_cell_dataset = image_dataset_from_directory(
    'Cell',
    color_mode = color_mode,
    validation_split=0.2,
    subset='training',
    seed=42,
    label_mode= 'categorical' ,
    image_size=image_size,
    batch_size=batch_size)

validation_cell_dataset = image_dataset_from_directory(
  'Cell',
  color_mode=color_mode,
  validation_split=0.2,
  subset='validation',
  seed=42,
  label_mode= 'categorical',
  image_size=image_size,
  batch_size=batch_size)

test_cell_dataset = image_dataset_from_directory(
  'test',
  shuffle=False,
  labels=None,
  color_mode=color_mode,
  image_size=image_size,
  batch_size=1)



print('Train label names:', train_cell_dataset.class_names)
print('Test label names (empty):', test_cell_dataset.class_names)
```

## Final model
A data augmentation layer was added and the model complexity was adjusted.

```python
from keras.layers.preprocessing.image_preprocessing import RandomContrast
from numpy.core.fromnumeric import reshape
from keras import activations
from keras.callbacks import EarlyStopping
from keras.constraints import MaxNorm
from keras.layers import Rescaling
from keras.models import Sequential
from keras.layers import Dense, Conv2D, Flatten,MaxPooling2D,Dropout,BatchNormalization,RandomFlip,RandomZoom,LayerNormalization
from sklearn.linear_model import LinearRegression
from tensorflow.keras.callbacks import ModelCheckpoint
from tensorflow.keras.callbacks import EarlyStopping



num_classes=len(train_cell_dataset.class_names)
print(num_classes)

model = Sequential([




Rescaling(1./255),
 RandomContrast(0.02,seed=60),
 RandomFlip( mode="horizontal_and_vertical", seed=None),
 RandomZoom(
    0.3,
    width_factor=None,
    fill_mode="reflect",
    interpolation="bilinear",
    seed=None,
    fill_value=0.0,

),


  Conv2D(32, kernel_size=(3,3), padding='same',activation="relu",data_format=None,input_shape=(36,36,1)),
  Conv2D(32, kernel_size=(3,3), padding='same',activation="relu"),
  Conv2D(32, kernel_size=(3,3), padding='same',activation="relu"),
  Conv2D(64, kernel_size=(3,3), padding='same',activation="relu"),
  Dropout(0.2),
  MaxPooling2D(pool_size=(2, 2)),


  Conv2D(64, kernel_size=(3,3),padding='same', activation="relu"),
  Conv2D(64, kernel_size=(3,3), padding='same',activation="relu"),
  Conv2D(128, kernel_size=(3,3),padding='same', activation="relu"),
  Conv2D(128, kernel_size=(3,3),padding='same', activation="relu"),
  MaxPooling2D(pool_size=(2, 2),strides=2),
  Dropout(0.2),
  Conv2D(128, kernel_size=(3,3),padding='same', activation="relu"),
  Conv2D(256, kernel_size=(3,3),padding='same', activation="relu"),
  MaxPooling2D(pool_size=(2, 2)),
  Dropout(0.2),

  Flatten(),

  Dropout(0.4),
  Dense(256, activation="relu"),
  Dense(512, activation="relu"),
  Dense(num_classes, activation="softmax"),








])

crossentropy = keras.losses.CategoricalCrossentropy()
learning_rate = 0.03
optimizer = tf.keras.optimizers.Adamax(learning_rate=learning_rate)
# optimizer = tf.keras.optimizers.Adamax()
accuracy = keras.metrics.CategoricalAccuracy()





# batch_size = 128
# epochs = 15

# # Set up checkpointing callback, called when training:
checkpoint_check=ModelCheckpoint(
    filepath='checkpoints',
    save_weights_only=True,
    monitor='val_accuracy',
    mode='max',
    save_best_only=True)

# checkpoint_lr = LinearRegression()



checkpoint_ear = EarlyStopping(   monitor='val_accuracy',
    min_delta=0,
    patience=25,
    verbose=0,
    mode='auto',
    baseline=None,
    restore_best_weights=True)

model.compile(loss="categorical_crossentropy", optimizer='Adamax', metrics=["accuracy"])



```

```python
history1=model.fit(train_cell_dataset, validation_data=validation_cell_dataset, epochs=50,callbacks=[checkpoint_ear,checkpoint_check])
```

```python
import matplotlib.pyplot as plt
%matplotlib inline

accuracy = history1.history['accuracy']
val_accuracy = history1.history['val_accuracy']
loss = history1.history['loss']
val_loss = history1.history['val_loss']
epochs = range(len(accuracy))

plt.plot(epochs, accuracy, 'bo', label='Training accuracy')
plt.plot(epochs, val_accuracy, 'b', label='Validation accuracy')
plt.title('Training and validation accuracy')
plt.legend()
plt.figure()

plt.plot(epochs, loss, 'bo', label='Training loss')
plt.plot(epochs, val_loss, 'b', label='Validation loss')
plt.title('Training and validation loss')
plt.legend()
plt.show()
```
